```python
import numpy as np
import sys
```

```python
sys.path.append("../sketchformer")
from basic_usage.sketchformer import continuous_embeddings
sketchformer = continuous_embeddings.get_pretrained_model()
```

```python
with open('../outputs/labels.txt', 'r') as f:
    classes = list(map(lambda s: s.strip(), f.readlines()))

for name in classes:
    print(name)
    data = np.load(
        f'../data/quickdraw/{name}.npz', encoding='latin1', allow_pickle=True)
    sketches = np.concatenate([data['train'], data['valid'], data['test']])
    print(len(sketches))
    preds = np.array(sketchformer.classify(sketches))
    filtered_sketches = sketches[preds == name]
    print(len(filtered_sketches))
    np.savez_compressed(f"../data/quickdraw_filtered/{name}.npz", sketches=filtered_sketches)
```
